```python
from pylab import *
from scipy import fft
from sklearn import svm
import numpy as np
```

```python
# remove the trailing empty line from all csv files before importing them!
sitting_csv = ["./data/sitting1.csv", "./data/sitting2.csv", "./data/sitting3.csv", "./data/sitting4.csv"]
sword_csv = ["./data/sword1.csv", "./data/sword2.csv", "./data/sword3.csv", "./data/sword4.csv"]
walking_csv = ["./data/walking1.csv", "./data/walking2.csv", "./data/walking3.csv", "./data/walking4.csv"]
```

```python
def read_data(filename):
    x = []
    y = []
    z = []
    avg = []
    for line in open(filename, "r").readlines():
        _x, _y, _z = map(int,line.strip().split(","))
        x.append(_x)
        y.append(_y)
        z.append(_z)
        avg.append((_x+_y+_z)/3)
    return avg
```

```python
sitting_raw = [read_data(f) for f in sitting_csv]
sword_raw  = [read_data(f) for f in sword_csv]
walking_raw  = [read_data(f) for f in walking_csv]
```

```python
# cut off 
all = sitting_raw + sword_raw + walking_raw
minlen = min([len(x) for x in all])
print "Cutting off after", minlen , "samples"
sitting_cut = [l[:minlen] for l in sitting_raw]
sword_cut = [l[:minlen] for l in sword_raw]
walking_cut = [l[:minlen] for l in walking_raw]
```

```python
sitting_freq = [np.abs(fft(l)/len(l))[1:len(l)/2] for l in sitting_cut]
sword_freq = [np.abs(fft(l)/len(l))[1:len(l)/2] for l in sword_cut]
walking_freq = [np.abs(fft(l)/len(l))[1:len(l)/2] for l in walking_cut]
```

```python
for x in range(4):
    plot(range(len(walking_freq[x])), walking_freq[x])
```

```python
SITTING = 0
SWORD = 1
WALKING = 2
categories = [SITTING] * 3 + [SWORD] * 3 + [WALKING] * 3
training_data = sitting_freq[:3] + sword_freq[:3] + walking_freq[:3]
print training_data
c.fit(training_data,categories)
```
