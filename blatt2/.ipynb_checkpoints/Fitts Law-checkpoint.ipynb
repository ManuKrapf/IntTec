```python
from pylab import *
%matplotlib inline
rcParams['figure.figsize'] = (16,9)
```

```python
data = genfromtxt("user1.csv", dtype=None, delimiter=";", names=True)
```

```python
data
```

```python
A = data['distance']
```

```python
W = data['widths']
```

```python
ID = log2(A/W +1)
```

```python
MT = data['time']
```

```python
scatter(ID, MT)
```

```python
scatter(ID, MT, alpha=0.2)
```

```python
b, a = polyfit(ID, MT, 1)
```

```python
a, b
```

```python
from scipy.stats import linregress
b,a, r_value, p_value, std_err = linregress(ID,MT)
```

```python
a, b, r_value, p_value, std_err
```

```python
xr = ID
```

```python
# create a straight line, i.e. corresponding y values at every given x value
yr = polyval([b,a],xr)
```

```python
plot(xr, yr)
scatter(ID, MT)
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(MT,yr)
```
